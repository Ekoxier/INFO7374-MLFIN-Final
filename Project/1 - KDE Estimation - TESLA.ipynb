```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Download data from Yahoo Finance
data = yf.download("TSLA", start="2021-01-01", end="2022-12-31")



# Calculate daily returns
returns = data['Adj Close']
median = returns.median()
mean = returns.mean()

# Plot actual distribution and KDE estimate
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(20, 20))
sns.lineplot(data=returns, color='red', markers=True,ax=axs[1])
sns.kdeplot(returns, shade=True, color='blue', ax=axs[0],bw_adjust=0.25,cut=0)
plt.axhline(median, color='green', linestyle='dashdot',label="Median")
plt.axhline(mean, color='black', linestyle="dashed",label="Mean")

plt.legend(prop={'size': 20})
axs[0].set_title('KDE Estimate')
axs[1].set_title('Actual Distribution')
plt.show()
```

## Observations:
- Multimodal distribution for Tesla, peak price in period = `$400`, and lowest price ~`$110`.
- Due to multimodal nature of KDE estimate, bandwidth parameter is lowered to capture different peaks.
- Mean value of the stock price during this period is `$275` and median value is `$250`.
